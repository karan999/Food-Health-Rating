```python
re = urllib.request.urlopen('http://ratings.food.gov.uk/open-data/en-GB').read()
```

```python
soup = BeautifulSoup(re , "lxml")
```

```python
open_data_static = soup.find('div' , {'id' : 'openDataStatic'})
```

```python
data_from_ods = open_data_static.find_all('td')
```

```python
open_data_static.find_all('h2')
```

## Extracting URL

```python
health_rating_url = {}
```

```python
for i in range(len(data_from_ods)):
    
    try:
        print (data_from_ods[i].a['title'], data_from_ods[i].a['href'])
        
        health_rating_url[data_from_ods[i].a['title']] = data_from_ods[i].a['href']
                
    except TypeError:
        
        pass
```

```python
health_rating_url
```

```python
health_rating_url['Aberdeen City: English language']
```

- Iterating through the dictionary to take each URL from the value of the dictionary and using this to extract the contents of each XML 
- Creating a new folder, if it doesn't exist, to store the saved xml files 

```python
for i,j in health_rating_url.items():
        
    folder = 'Data'
        
    try: 
        os.mkdir(folder)
        
    except FileExistsError: 
        
        pass
    
    path = folder + '\\' + i.replace(': English language', '') + '.xml'
        
    req = requests.get(j) # calling each url in the dictionary 
    
    req.encoding = 'utf-8' # changing encoding as it caused problems 
    
    file = req.text # extracting the text from the html 
    
    with open(path, 'w') as f: # writing each file to the 'Data' folder followed by their name
        
        f.write(file)
        
        f.close()
        
    print ("Created xml file for: %s" % i.replace(': English language' , ''))
```
